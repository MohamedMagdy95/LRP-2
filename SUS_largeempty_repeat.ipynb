```python
import random as rand
from deap import base, creator, tools, algorithms
import pandas as pd


import random as rand

# Custom crossover function to maintain unique values in chromosomes
def cxUnique(ind1, ind2):
    size = len(ind1)

    # Check if size is sufficient
    if size <= 14:
        raise ValueError("Individuals must be at least 15 elements long.")

    # Generate crossover points
    cxpoint1, cxpoint2 = sorted(rand.sample(range(1, size), 2))

    # Ensure crossover points are within bounds
    if not (0 <= cxpoint1 < size) or not (0 <= cxpoint2 < size):
        raise ValueError("Crossover points are out of bounds.")

    # Create temporary lists for crossover
    temp1_slice = ind1[cxpoint1:cxpoint2+1]
    temp2_slice = ind2[cxpoint1:cxpoint2+1]

    # Create unique individuals after crossover
    temp1 = temp1_slice + [item for item in ind2 if item not in temp1_slice]
    temp2 = temp2_slice + [item for item in ind1 if item not in temp2_slice]

    # Ensure indices 8 and 14 have the same values across the population
    if size > 14:
        if ind1[8] != ind2[8]:
            ind1[8] = ind2[8]
            ind2[8] = ind1[8]
        if ind1[14] != ind2[14]:
            ind1[14] = ind2[14]
            ind2[14] = ind1[14]

    # Fill the chromosomes to maintain the fixed length
    ind1[:] = temp1[:size]
    ind2[:] = temp2[:size]

    return ind1, ind2



def cusMut(ind, indpb):
    used_numbers = []
    used_numbers1 = []
    used_numbers2 = []
    temp1 = 0
    temp2 = 0
    temp3 = 0
    fixed_indices = [8, 14]  # Indices that should not be changed
    for a in range(len(ind)):
      used_numbers1.append(ind[a])

    for b in range(len(ind)):
      if b == 8 or b == 14 or ind[b] == 1:
        continue
      else:
        temp3 = ind[b]
        for c in range(len(ind)):
          if ind[b] == ind[c] and b < 10 and ind[b] != 1:
            while temp3 in used_numbers1:
              temp3 = rand.randrange(6,16)
            used_numbers1.append(temp3)
            ind[b] = temp3
          elif ind[b] == ind[c] and b >= 10 and ind[b] != 1:
            while temp3 in used_numbers1:
              temp3 = rand.randrange(16,26)
            used_numbers1.append(temp3)
            ind[b] = temp3

    for a in range(len(ind)):
      used_numbers.append(ind[a])

    print("before mutation",ind)
    for i in range(len(ind)):
        # print(i, "ind")
        temp1 = 0
        temp2 = 0
        if i in fixed_indices:
            continue  # Skip mutation for fixed indices

        if rand.random() < indpb and ind[i] != 1 and i < 10:
          temp1 = rand.randrange(6, 16)
          while temp1 in used_numbers:
            temp1 = rand.randrange(6,16)
          used_numbers.append(temp1)
          ind[i] = temp1  # Mutate with a value from 1 to 24
        # if i == 8 or i == 14 and rand.random() < indpb:
        #   ind[8]=ind[14]

        if rand.random() < indpb and ind[i] != 1 and i >= 10:
          temp2 = rand.randrange(16, 26)
          while temp2 in used_numbers:
            temp2 = rand.randrange(16,26)
          used_numbers.append(temp2)
          ind[i] = temp2

        for a in range(len(ind)):
          used_numbers2.append(ind[a])

        for b in range(len(ind)):
          if b == 8 or b == 14 or ind[b] == 1:
            continue
          else:
            temp3 = ind[b]
        for c in range(len(ind)):
          if ind[b] == ind[c] and b < 10 and ind[b] != 1:
            while temp3 in used_numbers2:
              temp3 = rand.randrange(6,16)
            used_numbers2.append(temp3)
            ind[b] = temp3
          elif ind[b] == ind[c] and b >= 10 and ind[b] != 1:
            while temp3 in used_numbers2:
              temp3 = rand.randrange(16,26)
            used_numbers2.append(temp3)
            ind[b] = temp3

    print("after mutation",ind)
    return ind

def filter_invalid(pop):
    valid_pop = []

    # Iterate over each individual (chromosome) in the population
    for ind in pop:
        usednumbers = []

        # Validate the chromosome
        if 0 in ind:
            continue
        if len(set(ind)) != len(ind) - ind.count(1) + 1 or len(ind) != 18:
            continue

        for a in range(len(ind)):
            usednumbers.append(ind[a])

        for i in range(len(ind)):
            if i < 10:  # For indices 0 to 9
                if i == 8 or ind[i] == 1:
                    continue  # Skip index 8 and value 1
                if not (6 <= ind[i] <= 15):
                    temp1 = rand.randrange(6, 16)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(6, 16)
                    usednumbers.append(temp1)
                    ind[i] = temp1

            else:  # For indices 10 to 17
                if i == 14 or ind[i] == 1:
                    continue  # Skip index 14 and value 1
                if not (16 <= ind[i] <= 25):
                    temp1 = rand.randrange(16, 26)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(16, 26)
                    usednumbers.append(temp1)
                    ind[i] = temp1

        # Ensure indices 8 and 14 are the same
        if ind[8] != ind[14]:
            ind[8] = ind[14]

        # Ensure specific indices are set to 1
        if ind[0] != 1 or ind[4] != 1 or ind[7] != 1 or ind[9] != 1 or ind[13] != 1 or ind[17] != 1:
            ind[0] = 1
            ind[4] = 1
            ind[7] = 1
            ind[9] = 1
            ind[13] = 1
            ind[17] = 1

        # Check capacity constraints
        if (capacitymatrix[ind[1]-1][1] + capacitymatrix[ind[2]-1][1] + capacitymatrix[ind[3]-1][1] <= 30 and
            capacitymatrix[ind[10]-1][1] + capacitymatrix[ind[11]-1][1] + capacitymatrix[ind[12]-1][1] <= 30):
            valid_pop.append(ind)
        else:
            for j in range(len(ind)):
                if j == 0 or j == 4 or j == 7 or j == 9 or j == 13 or j == 17:
                    ind[j] = 1
                if j == 8:
                    temp1 = rand.randrange(34, 38)
                    ind[8] = temp1
                    ind[14] = temp1

                if ind[j] == 1:
                    continue

                if j in range(1, 4):  # For indices 1, 2, 3
                    temp1 = rand.randrange(6, 16)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(6, 16)
                    usednumbers.append(temp1)
                    ind[j] = temp1
                elif j in range(10, 13):  # For indices 10, 11, 12
                    temp1 = rand.randrange(16, 26)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(16, 26)
                    usednumbers.append(temp1)
                    ind[j] = temp1

            # Re-check capacity constraints after modifications
            if (capacitymatrix[ind[1]-1][1] + capacitymatrix[ind[2]-1][1] + capacitymatrix[ind[3]-1][1] <= 30 and
                capacitymatrix[ind[10]-1][1] + capacitymatrix[ind[11]-1][1] + capacitymatrix[ind[12]-1][1] <= 30):
                valid_pop.append(ind)

    return valid_pop


import random as rand
from deap import base, creator, tools, algorithms
import pandas as pd

# Custom crossover function to maintain unique values in chromosomes
def cxUnique(ind1, ind2):
    size = len(ind1)

    # Check if size is sufficient
    if size <= 14:
        raise ValueError("Individuals must be at least 15 elements long.")

    # Generate crossover points
    cxpoint1, cxpoint2 = sorted(rand.sample(range(1, size), 2))

    # Ensure crossover points are within bounds
    if not (0 <= cxpoint1 < size) or not (0 <= cxpoint2 < size):
        raise ValueError("Crossover points are out of bounds.")

    # Create temporary lists for crossover
    temp1_slice = ind1[cxpoint1:cxpoint2+1]
    temp2_slice = ind2[cxpoint1:cxpoint2+1]

    # Create unique individuals after crossover
    temp1 = temp1_slice + [item for item in ind2 if item not in temp1_slice]
    temp2 = temp2_slice + [item for item in ind1 if item not in temp2_slice]

    # Ensure indices 8 and 14 have the same values across the population
    if size > 14:
        if ind1[8] != ind2[8]:
            ind1[8] = ind2[8]
            ind2[8] = ind1[8]
        if ind1[14] != ind2[14]:
            ind1[14] = ind2[14]
            ind2[14] = ind1[14]

    # Fill the chromosomes to maintain the fixed length
    ind1[:] = temp1[:size]
    ind2[:] = temp2[:size]

    return ind1, ind2

def cusMut(ind, indpb):
    used_numbers = []
    used_numbers1 = []
    used_numbers2 = []
    temp1 = 0
    temp2 = 0
    temp3 = 0
    fixed_indices = [8, 14]  # Indices that should not be changed
    for a in range(len(ind)):
      used_numbers1.append(ind[a])

    for b in range(len(ind)):
      if b == 8 or b == 14 or ind[b] == 1:
        continue
      else:
        temp3 = ind[b]
        for c in range(len(ind)):
          if ind[b] == ind[c] and b < 10 and ind[b] != 1:
            while temp3 in used_numbers1:
              temp3 = rand.randrange(6,16)
            used_numbers1.append(temp3)
            ind[b] = temp3
          elif ind[b] == ind[c] and b >= 10 and ind[b] != 1:
            while temp3 in used_numbers1:
              temp3 = rand.randrange(16,26)
            used_numbers1.append(temp3)
            ind[b] = temp3

    for a in range(len(ind)):
      used_numbers.append(ind[a])

    # print("before mutation",ind)
    for i in range(len(ind)):
        # print(i, "ind")
        temp1 = 0
        temp2 = 0
        if i in fixed_indices:
            continue  # Skip mutation for fixed indices

        if rand.random() < indpb and ind[i] != 1 and i < 10:
          temp1 = rand.randrange(6, 16)
          while temp1 in used_numbers:
            temp1 = rand.randrange(6,16)
          used_numbers.append(temp1)
          ind[i] = temp1  # Mutate with a value from 1 to 24
        # if i == 8 or i == 14 and rand.random() < indpb:
        #   ind[8]=ind[14]

        if rand.random() < indpb and ind[i] != 1 and i >= 10:
          temp2 = rand.randrange(16, 26)
          while temp2 in used_numbers:
            temp2 = rand.randrange(16,26)
          used_numbers.append(temp2)
          ind[i] = temp2

        for a in range(len(ind)):
          used_numbers2.append(ind[a])

        for b in range(len(ind)):
          if b == 8 or b == 14 or ind[b] == 1:
            continue
          else:
            temp3 = ind[b]
        for c in range(len(ind)):
          if ind[b] == ind[c] and b < 10 and ind[b] != 1:
            while temp3 in used_numbers2:
              temp3 = rand.randrange(6,16)
            used_numbers2.append(temp3)
            ind[b] = temp3
          elif ind[b] == ind[c] and b >= 10 and ind[b] != 1:
            while temp3 in used_numbers2:
              temp3 = rand.randrange(16,26)
            used_numbers2.append(temp3)
            ind[b] = temp3

    # print("after mutation",ind)
    return ind

def filter_invalid(pop):
    valid_pop = []

    # Iterate over each individual (chromosome) in the population
    for ind in pop:
        usednumbers = []

        # Validate the chromosome
        if 0 in ind:
            continue
        if len(set(ind)) != len(ind) - ind.count(1) + 1 or len(ind) != 18:
            continue

        for a in range(len(ind)):
            usednumbers.append(ind[a])

        for i in range(len(ind)):
            if i < 10:  # For indices 0 to 9
                if i == 8 or ind[i] == 1:
                    continue  # Skip index 8 and value 1
                if not (6 <= ind[i] <= 15):
                    temp1 = rand.randrange(6, 16)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(6, 16)
                    usednumbers.append(temp1)
                    ind[i] = temp1

            else:  # For indices 10 to 17
                if i == 14 or ind[i] == 1:
                    continue  # Skip index 14 and value 1
                if not (16 <= ind[i] <= 25):
                    temp1 = rand.randrange(16, 26)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(16, 26)
                    usednumbers.append(temp1)
                    ind[i] = temp1

        # Ensure indices 8 and 14 are the same
        if ind[8] != ind[14]:
            ind[8] = ind[14]

        # Ensure specific indices are set to 1
        if ind[0] != 1 or ind[4] != 1 or ind[7] != 1 or ind[9] != 1 or ind[13] != 1 or ind[17] != 1:
            ind[0] = 1
            ind[4] = 1
            ind[7] = 1
            ind[9] = 1
            ind[13] = 1
            ind[17] = 1

        # Check capacity constraints
        if (capacitymatrix[ind[1]-1][1] + capacitymatrix[ind[2]-1][1] + capacitymatrix[ind[3]-1][1] <= 30 and
            capacitymatrix[ind[10]-1][1] + capacitymatrix[ind[11]-1][1] + capacitymatrix[ind[12]-1][1] <= 30):
            valid_pop.append(ind)
        else:
            for j in range(len(ind)):
                if j == 0 or j == 4 or j == 7 or j == 9 or j == 13 or j == 17:
                    ind[j] = 1
                if j == 8:
                    temp1 = rand.randrange(34, 38)
                    ind[8] = temp1
                    ind[14] = temp1

                if ind[j] == 1:
                    continue

                if j in range(1, 4):  # For indices 1, 2, 3
                    temp1 = rand.randrange(6, 16)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(6, 16)
                    usednumbers.append(temp1)
                    ind[j] = temp1
                elif j in range(10, 13):  # For indices 10, 11, 12
                    temp1 = rand.randrange(16, 26)
                    while temp1 in usednumbers:
                        temp1 = rand.randrange(16, 26)
                    usednumbers.append(temp1)
                    ind[j] = temp1

            # Re-check capacity constraints after modifications
            if (capacitymatrix[ind[1]-1][1] + capacitymatrix[ind[2]-1][1] + capacitymatrix[ind[3]-1][1] <= 30 and
                capacitymatrix[ind[10]-1][1] + capacitymatrix[ind[11]-1][1] + capacitymatrix[ind[12]-1][1] <= 30):
                valid_pop.append(ind)

    return valid_pop


# Create the toolbox with the right settings
toolbox = base.Toolbox()

# Define the individual and population
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))  # Minimize the fitness function
creator.create("Individual", list, fitness=creator.FitnessMin)  # Create individual

# Register the initial individual creation method
toolbox.register("individual", lambda: creator.Individual(rand.choices(range(1, 19), k=7)))

# Define a wrapper function for population creation to avoid the error
def create_population(n):
    return initial_population(n)

# Register the initial population creation method
toolbox.register("population", create_population)

# Register the genetic operators
toolbox.register("evaluate", fitness_func1)  # Evaluation function (fitness function)
toolbox.register("mate", tools.cxOnePoint)  # Custom crossover operation
toolbox.register("mutate", cusMut, indpb=0.45)  # Mutation operation
toolbox.register("select", tools.selStochasticUniversalSampling)  # Change to SUS selection

# Create the initial population
population = toolbox.population(n=pop_size)

# Set the parameters for the genetic algorithm
ngen = 900  # Number of generations
cxpb = 0.90  # Probability of crossover
mutpb = 0.10  # Probability of mutation
num_offspring = 5  # Number of offspring to produce each generation
num_parents = 5  # Number of parents to select
hof = tools.HallOfFame(1)
# Define the custom steady-state evolutionary algorithm
def custom_steady_state(population, toolbox, cxpb, mutpb, ngen, num_offspring, stats=None, halloffame=hof, verbose=__debug__):
    logbook = tools.Logbook()
    logbook.header = ['gen', 'nevals'] + (stats.fields if stats else [])

    best_fitness_values = []

    # Evaluate the entire population
    # print("Evaluating initial population...")
    invalid_ind = [ind for ind in population if not ind.fitness.valid]
    fitnesses = map(toolbox.evaluate, invalid_ind)
    for ind, fit in zip(invalid_ind, fitnesses):
        ind.fitness.values = fit
    # print(f"Initial population evaluated. Invalid individuals: {len(invalid_ind)}")

    if hof is not None:
        hof.update(population)

    record = stats.compile(population) if stats else {}
    logbook.record(gen=0, nevals=len(invalid_ind), **record)
    if verbose:
        print(logbook.stream)

    for gen in range(1, ngen + 1):
        # print(f"Generation {gen}/{ngen}...")
        for _ in range(num_offspring):
            # Select parents
            parents = toolbox.select(population, num_parents)
            parent1 = toolbox.clone(parents[0])
            parent2 = toolbox.clone(parents[1])
            parent3 = toolbox.clone(parents[2])
            parent4 = toolbox.clone(parents[3])
            parent5 = toolbox.clone(parents[4])

            # Apply crossover and mutation
            if rand.random() < cxpb:
                toolbox.mate(parent1, parent2)
                del parent1.fitness.values
                del parent2.fitness.values

            if rand.random() < mutpb:
                toolbox.mutate(parent1)
                toolbox.mutate(parent2)
                del parent1.fitness.values
                del parent2.fitness.values

            # Evaluate the new individuals
            invalid_ind = [ind for ind in [parent1, parent2] if not ind.fitness.valid]
            if invalid_ind:
                fitnesses = map(toolbox.evaluate, invalid_ind)
                for ind, fit in zip(invalid_ind, fitnesses):
                    ind.fitness.values = fit

            # Filter out invalid individuals
            valid_offspring = filter_invalid([parent1, parent2])
            if valid_offspring:
                worst_inds = tools.selWorst(population, len(valid_offspring))
                for worst, new in zip(worst_inds, valid_offspring):
                    population.remove(worst)
                    population.append(new)

        if hof is not None:
            hof.update(population)

        record = stats.compile(population) if stats else {}
        logbook.record(gen=gen, nevals=len(invalid_ind), **record)

        best_individual = tools.selBest(population, k=1)[0]
        final_best_individual = tools.selBest(population, k=1)[0]
        best_fitness_values.append(best_individual.fitness.values[0])
        # print(f"Best fitness at generation {gen}: {best_individual.fitness.values[0]}")
    print("Final Best Chromosome:")
    print("Chromosome:", final_best_individual)
    print("Fitness Value:", final_best_individual.fitness.values[0])
    return population, logbook, best_fitness_values
```
